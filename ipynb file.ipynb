# For Loops

Python for loops are used to loop through an iterable object (like a list, tuple, set, etc.) and perform the same action for each value. 

For example, a for loop would allow us to iterate through a list, performing the same action (or block of code) on each item in the list.


Personally, I love for loops and I use them all the time and they have a lot of different use cases - which we show in this video as well as future videos.

Here is the basics flow of a for loop:


```python
#The syntax looks something like this:

for val in sequence(List, str, etc):
    [body of code to run on that value]
    
#The value can be named anything variable name we want
    
#So it's somewhat like an if statement in structure kind of - except we are assigning a value to each looped value

if x > y:
    [body of code to run on that value]
```

```python
# Let's start by creating a simple list


integers = [1,2,3,4,5]
```

```python
# Let's look at the syntax - we have the for keyword which indicates a for loop - then we have in

#Again, we can name that variable anything we want

for number in integers:
    print(number)
    
#Again, we can name that variable anything we want
```

```python
# Now we choose to return the values in the list, but we don't have to, we can really do anything we want. 
# Like in this example we are really only using the values as almost like a counter

for number in integers:
    print('Hello!')
```

```python
#We can also perform calculations in the body of code - you can really do anything in here for each value in the sequence

integers = [1,2,3,4,5]

for num in integers:
    power_of_num = num**num
    print(power_of_num)
```

```python

sale_amount = [1,20.70,35.2,50,100,125,70.59,106.21,700]

small_sale = []
medium_sale = []
large_sale = []


for sale in sale_amount:
    if sale < 50:
        small_sale.append(sale)
    elif sale < 100:
        medium_sale.append(sale)
    else:
        large_sale.append(sale)
    

#Now we can print these out and see what we have - each sale is put into the correct lists
print('Small Sale: ', small_sale)
print('Medium Sale: ', medium_sale)
print('Large Sale: ', large_sale)
```

```python
#Now let's look at enumeration in python - this basically means adding a counter to your output when using a loop

flavor_list = ['Mint Chocolate Chip','Chocolate','Fudge Ripple','Cookie Dough']
i = 0

for flavor in flavor_list:
    print(i,flavor)
    i += 1
    
#So each time this loops through - is adding 1
```

```python
# There's also an enumerate function that'll do this for you:
#Python's enumerate() lets you write Pythonic for loops when you need a count and the value from an iterable.
flavor_list

for counter, flavor in enumerate(flavor_list):
     print(counter, flavor)
```

```python
# Now let's look at looping through a dictionary



customer_dict = {'name': 'John Smith', 'amount spent': 75, 'items purchased': ['Shovel', 'Dirt']}


for values in customer_dict.values():
    print(values)
```

```python
#We can also look at both the key and the value by using.items()

for key, value in customer_dict.items():
    print(key, '->',value)
```

```python
#Let's look at one more use Case of a For Loop - we want to find numbers that are divisible by 3, then put it in a list

divis_by_three = []

for num in range(100):
    if num % 3 == 0:
        divis_by_three.append(num)

#Now we want to write an if/else statement that will tell us who to let in and who not to let in
print(divis_by_three)
```
